# ANALISIS DE PRODUCCIÓN EN POZOS NO CONVENCIONALES

## DATA STORY TELLING

Los hidrocarburos( petróleo y/o gas)  se alojan en diferentes tipos de rocas y justamente el tipo de roca determina el tipo de yacimiento, Convencional o No Convencional. Los yacimientos convencionales son aquellos en donde el hidrocarburo está alojado en la “Roca Reservorio”, es decir fue generado en otro tipo de roca y migro hasta quedar depositado o "entrampado". Esta “Roca Reservorio” tiene características particulares que permiten la extracción de los hidrocarburos, como son su permeabilidad y porosidad. La permeabilidad tiene que ver con la capacidad de la roca de permitir la fluidez del fluido circulante y la porosidad es la cantidad de poros interconectados disponibles para almacenar el hidrocarburo. Por su parte los yacimientos no convencionales son aquellos en donde el hidrocarburo se almaceno directamente en la “Roca Generadora” o una roca de baja permeabilidad, que jusatmente debido a esa baja permeabilidad "entrampo" el hidrocarburo y este no continuo migrando. Para cada caso se necesitan tecnologías diferentes para que dichos hidrocarburos puedan ser extraidos.

Hasta hace algunos años, no se había desarrollado la explotación no convencional, debido a que la explotacion se enfocaba principalmente en yacimientos convencionales la inversion es menor y la misma se recupera mas rapido puesto que el hidrocarburo tiende a migrar hacia la superficie y solo es impedido por la Roca Sello por sus características constructivas; pero en un yacimiento no convencional la misma Roca Madre o Generadora actúa como un propio sello, y por supuiesto esto significa que el hidrocarburo no fluye tan facilmiente y por lo tanto no es tan fácil extraerlo. Se necesitan tecnologías más avanzadas, como la inyección de grandes cantidades de agua, arena y aditivos químicos, en perforaciones horizontales, a profundidades mayores a 3500 metros, a presiones extremas.

Algunos años atrás comenzaron a descubrirse estos nuevos yacimientos no convenvionales  que obligaron a la industria del Petróleo y el Gas a recapacitarse y descubrir nuevas tecnologías para la extracción del hidrocarburo. El interes surge principalmente de dos razones: para la formación de un yacimiento de tipo no convencional los requisitos son mucho menos que los de uno convencional, debido a que es la misma roca generadora la que lo contiene; y que además hay muchos más yacimientos de este tipo que aún no fueron explorados, pudiendo superar ampliamente las reservas de yacimientos convencionales. Es por esto ultimo principalmente que a pesar de que la extración en estos tipos de yacimientos requiere una inversion importante con alto riesgo, las industrias deciden igualmente invertir en este tipo de proyectos.

Una de las tecnicas mas comunmente usadas para la extraccion del hidrocarburo en yacimientos no convencionales es el fracturamiento hidraulico o "fracking". Como mencionamos anteriormente, el hidrocarburo de un yacimiento no convencional está almacenado en un reservorio sellado (no existe suficiente permealibidad para que fluya), por lo que necesita de cierta estimulación para migrar hacia la superficie, y eso se lleva a cabo mediante fracturas hidraúlicas que generen canales para que pueda ascender y fluir. Además, en función de la dificultad de migración hacia la superficie, se desarrollan métodos de extracción “horizontal” con el fin de discurrir, al contrario de un pozo convencional que por lo general se perfora de manera vertical.

Este método empezó a aplicarse industrialmente en 1949 en los Estados Unidos y se estima que en 2010 esta técnica estaba presente en aproximadamente el 60 % de los pozos de extracción en uso y cabe destacar que hoy en día, la estimulación hidraúlica ha mejorado notablemente el rendimiento de los pozos y su tiempo de vida en servicio. En Argentina se comenzó a usar esta técnica en la formación Vaca Muerta, ubicada en la provincia de Neuquén, a partir del año 2011. Tambien ha sido usada en Santa Cruz, en cercanías a Los Perales. Argentina ocupa a nivel mundial el segundo puesto en recursos de gas no convencional y el cuarto en petróleo.
